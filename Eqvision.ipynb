```python
import seaborn as sns
import matplotlib.pyplot as plt


df = df[['latitude', 'longitude', 'depth', 'nst', 'horizontalError', 'mag']]

# Select the numerical columns for correlation analysis
numeric_cols = ['latitude', 'longitude', 'depth', 'mag', 'nst', 'horizontalError']

# Create correlation matrix
corr_matrix = df[numeric_cols].corr()

# Plot heatmap
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()

```

```python
x = df[['latitude', 'longitude', 'depth', 'nst', 'horizontalError']]
y = df['mag']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score


X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)


rf = RandomForestRegressor(n_estimators=100, random_state=42)

rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)


mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean squared error: ", mse)
print("R-squared score: ", r2)
```

```python
import numpy as np
import pandas as pd


latitude = 19.2543334960938
longitude = -155.410827636719
depth = 31.3199996948242
nst = 41
horizontalError = 0.66


feature_names = ["latitude", "longitude", "depth", "nst", "horizontalError"]



new_data = pd.DataFrame([[latitude, longitude, depth, nst, horizontalError]], columns=feature_names)


predicted_mag = rf.predict(new_data)

if predicted_mag<5.0:
  imp='Light'
elif predicted_mag>=5.0 and predicted_mag<6:
  imp='Moderate'
elif predicted_mag>=6 and predicted_mag<9.9:
  imp='Strong'
print(f"The predicted magnitude of the earthquake using the Random Forest model is: { predicted_mag[0]:.2f} and the earthquake is {imp}")
```
